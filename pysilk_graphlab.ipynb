```python
import silk
import pandas     as pd
import graphlab   as gl
import matplotlib as plt
from socket import inet_ntoa,inet_aton
from struct import unpack,pack

%matplotlib inline
if (not silk.site.have_site_config()):
    silk.site.init_site("samples/silk.conf")
```

```python

def ntoa(n):
    return inet_ntoa( pack("!L", n ))
    
def aton(a):
    return unpack("!L",inet_aton(a))[0]

def project_rec(rwrec):
    return{
        "sip":  str(rwrec.sip),
        "sint": int(rwrec.sip),
        "dip":  str(rwrec.dip),
        "dint": int(rwrec.dip),
        "bytes": rwrec.bytes
    }

def project_vertex(ip):
    return gl.Vertex(aton(ip),{'ip': ip})
    
def project_edge(rec):
    return gl.Edge(aton(rec[0]), aton(rec[1]), {'bytes': rec[2]})
```

```python
recs = map(project_rec,silk.silkfile_open("samples/1mil.rwz",silk.READ))
df   = pd.DataFrame.from_records(recs)
```

```python
records = df[['sip','dip','bytes']].groupby(['sip','dip'],as_index=False).sum()
vertices = map(project_vertex,df['sip'].unique())
edges    = map(project_edge, records.get_values())
```

```python
g=gl.SGraph()
g=g.add_vertices(vertices)
g=g.add_edges(edges)
```

```python
g.get_neighborhood(aton('119.165.228.80')).show()
```

```python
pr = gl.pagerank.create(g)
ranked=pr['pagerank']
```

```python
ranked.add_column(ranked.apply(lambda x: ntoa(x['__id'])))
```
